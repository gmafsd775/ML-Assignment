```python
import numpy as np
import pandas as pd
import matplotlib as plt
from sklearn import linear_model
```

```python
from google.colab import files
import pandas as pd

df = pd.read_csv("Salary_Data.csv")
df
```

```python
plt.xlabel('YearsExperience', color = "red")
plt.ylabel('Salary', color = "green")
plt.scatter(df.YearsExperience,df.Salary)
```

```python
lr = linear_model.LinearRegression()
lr.fit(df[['YearsExperience']],df.Salary)
```

```python
lr.predict([[35.0]])
```

```python
lr.score(df[['YearsExperience']],df.Salary)
```

```python
lr.coef_
```

```python
lr.intercept_
```
